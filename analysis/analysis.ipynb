```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bone_cancer = pd.read_csv('bone_cancer_prioritized.csv')
breast_cancer = pd.read_csv('breast_cancer_prioritized.csv')
lung_cancer = pd.read_csv('lung_cancer_prioritized.csv')
pancreatic_cancer = pd.read_csv('pancreatic_cancer_prioritized.csv')
```

## Bone Cancer

```python
sns.set_theme(style="whitegrid")
sns.set(font_scale=1.25)
g = sns.catplot(
    data=bone_cancer, kind="bar",
    x="Thresholded", y="Count", palette="Paired", alpha=.8, height=10
)
g.despine(left=True)
plt.xticks(rotation=90)
g.set(xlabel="Gene ID", ylabel="Instances in Top Fifth Percentile", title="Thresholded Gene Instances for Bone Cancer")
g.savefig('bone.png')
```

## Breast Cancer

```python
sns.set_theme(style="whitegrid")
sns.set(font_scale=1.25)
g = sns.catplot(
    data=breast_cancer, kind="bar",
    x="Thresholded", y="Count", palette="Paired", alpha=.8, height=10
)
g.despine(left=True)
plt.xticks(rotation=90)
g.set(xlabel="Gene ID", ylabel="Instances in Top Fifth Percentile", title="Thresholded Gene Instances for Breast Cancer")
g.savefig('boob.png')
```

## Lung Cancer

```python
sns.set_theme(style="whitegrid")
sns.set(font_scale=1.25)
g = sns.catplot(
    data=lung_cancer, kind="bar",
    x="Thresholded", y="Count", palette="Paired", alpha=.8, height=10
)
g.despine(left=True)
plt.xticks(rotation=90)
g.set(xlabel="Gene ID", ylabel="Instances in Top Fifth Percentile", title="Thresholded Gene Instances for Lung Cancer")
g.savefig('smokes.png')
```

## Pancreatic Cancer

```python
sns.set_theme(style="whitegrid")
sns.set(font_scale=1.25)
g = sns.catplot(
    data=pancreatic_cancer, kind="bar",
    x="Thresholded", y="Count", palette="Paired", alpha=.8, height=10
)
g.despine(left=True)
plt.xticks(rotation=90)
g.set(xlabel="Gene ID", ylabel="Instances in Top Fifth Percentile", title="Thresholded Gene Instances for Pancreatic Cancer")
g.savefig('venom.png')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(45, 25), sharey=True)
sns.set_theme(style="whitegrid")
fig.suptitle("Thresholded Gene Instances\n\n", size=45)
# Bone Cancer Plot
g1 = sns.barplot(x="Thresholded", y="Count", palette="Paired", data=bone_cancer, ax=axes[0, 0], alpha=.8)
g1.set(xlabel=None)
g1.set_ylabel("Instances in Top Fifth Percentile", size=35)
g1.set_title("Bone Cancer\n", size=35)
for item in g1.get_xticklabels():
    item.set_fontsize(25)
    item.set_rotation(90)

# Breast Cancer Plot
g2 = sns.barplot(x="Thresholded", y="Count", palette="Paired", data=breast_cancer, ax=axes[0, 1], alpha=.8)
g2.set(xlabel=None)
g2.set(ylabel=None)
g2.set_title("Breast Cancer\n", size=35)
for item in g2.get_xticklabels():
    item.set_fontsize(25)
    item.set_rotation(90)

# Lung Cancer Plot
g3 = sns.barplot(x="Thresholded", y="Count", palette="Paired", data=lung_cancer, ax=axes[1, 0], alpha=.8)
for item in g3.get_xticklabels():
    item.set_fontsize(25)
    item.set_rotation(90)
g3.set_ylabel("Instances in Top Fifth Percentile", size=35)
g3.set_xlabel("Gene ID", size=35)
g3.set_title("Lung Cancer\n", size=35)
# Pancreatic Cancer Plot
g4 = sns.barplot(x="Thresholded", y="Count", palette="Paired", data=pancreatic_cancer, ax=axes[1, 1], alpha=.8)
g4.set(ylabel=None)
for item in g4.get_xticklabels():
    item.set_fontsize(25)
    item.set_rotation(90)
g4.set_xlabel("Gene ID", size=35)
g4.set_title("Pancreatic Cancer\n", size=35)
plt.tight_layout()
plt.savefig('plots-tight.png')
```
